```python
s = "abcdefghijklmnopqrstuvwxyz"

# 前から
print(s[0]) # a
print(s[1]) # b
print(s[25]) # z

# 後から
print(s[-1]) # z
print(s[-3]) # x
print(s[-26]) # a

s = "abcdefghijklmnopqrstuvwxyz"

# 前から
print(s[0:7])
print(s[21:26])

# 後から
print(s[-5:-1])

s = "abcdefghijklmnopqrstuvwxyz"

print(s[:10]) # 開始位置を省略
print(s[10:]) # 終了位置を省略
print(s[:])   # 両方省略

s = "abcdefghijklmnopqrstuvwxyz"

print(s[0:10:2]) # 2個おき
print(s[1:20:3]) # 3個おき
print(s[-1:-10:-2]) # 後から

s = "にんてんどううどんてんに"
if s == s[::-1]: # 元の値とひっくり返した値が同値である
    print(f"「{s}」は回文です！")
else:
    print(f"「{s}」は回文ではないです。")

s = "Pythonは便利だが遅い"
if s == s[::-1]:
    print(f"「{s}」は回文です！")
else:
    print(f"「{s}」は回文ではないです。")

s = "no devil lived on"
if s == s[::-1]:
    print(f"「{s}」は回文です！")
else:
    print(f"「{s}」は回文ではないです。")

s = "TechAcademy"
for char in s:
    print(char)

s = "TechAcademy"
for char in s:
    print(char.upper())  # 各文字を大文字に変換して出力

s = "TechAcademy"
vowels = "aeiouAEIOU"  # 母音のリスト
for char in s:
    if char in vowels:
        print(char)

s = "TechAcademy"
length = len(s)
print(f"文字列の長さ: {length}")

s = "Hello, world!"
if len(s) > 10:
    print("10文字を超える長い文字列です。")
else:
    print("10文字以内の短い文字列です。")

s = ""
if len(s) == 0:
    print("文字列は空です。")
else:
    print("文字列は空ではありません。")
```
